```python
%matplotlib inline
from nuinsseg import NuInsSeg
import os
dataroot = os.path.join(os.environ['HOME'], 'datasets/nuscenes')
nuseg = NuInsSeg(version='v1.0-mini', dataroot=dataroot, verbose=True)
```

## Category histogram statistics

```python
nuseg.list_categories_nuseg()
```

## Render instance mask in image
- nuseg.render_2d_annotation(sample, sample_annotation_token)
- a sample annotation may correspond to more than one instance masks in different cameras.

```python

my_sample = nuseg.sample[44]
# my_sample
my_sample_annotation_token = my_sample['anns'][32]
my_sample_annotation = nuseg.get('sample_annotation', my_sample_annotation_token)

nuseg.render_2d_annotation(my_sample, my_sample_annotation_token)
```

## Render all the masks in the image

```python
CAM_SENSOR = ['CAM_FRONT', 'CAM_FRONT_RIGHT', 'CAM_BACK_RIGHT', 'CAM_BACK', 'CAM_BACK_LEFT', 'CAM_FRONT_LEFT']
POINT_SENSOR = 'LIDAR_TOP'

for cam in CAM_SENSOR:
    nuseg.render_2d_image(my_sample['data'][cam])
```
